**INTRODUCTION**

This project is all about the Real State market trends of the California city bases on the California housing dataset.The California Housing Prices Project is a comprehensive study focused on analyzing the trends, drivers, and implications of the housing market in the diverse and dynamic state of California as due to its geographical factors

**6. Write a programming construct (create a user defined function) to calculate the median value of the data set wherever required.**

```python
def medianValue(data):
    """
    Calculates the median value of a given dataset.
    """
    # Firstly sorting the data in ascending order
    sorted_data = sorted(data)

    # getting num. of datas are there
    length = len(sorted_data)

    # Calculating the median based on the number of datas
    if length % 2 == 0:
        # If the number of data is even, we will take the average of the two middle values
        mid_index_1 = (length // 2) - 1
        mid_index_2 = length // 2

        #average of both the middle values
        median = (sorted_data[mid_index_1] + sorted_data[mid_index_2]) / 2
    else:
        # If the number of data points is odd, take the middle value as median
        mid_index = length // 2
        median = sorted_data[mid_index]

    return median
```

```python
## calling function with total_bedrooms column
median = medianValue(fixed_data['total_bedrooms'])
print(median)
```

```python
## verifying with the built-in median method
fixed_data['total_bedrooms'].median()
```

**10. Please create a new column named total_bedroom_size. If the total bedrooms is 10 or less, it should be quoted as small. If the total bedrooms is 11 or more but less than 1000, it should be medium, otherwise it should be considered large.**

```python
## new_data data_set is the set i have created for displaying the data with no missing values
fixed_data.head()
```

```python
# defining conditions
cases = [
    fixed_data['total_bedrooms'] <= 10,
    (fixed_data['total_bedrooms'] >= 11) & (fixed_data['total_bedrooms'] < 1000),
    fixed_data['total_bedrooms'] >= 1000
]

#defining choicelist
room_size = ["SMALL", "MEDIUM", "LARGE" ]

# copying the dataset to avoid getting error
copied_set = fixed_data.copy()

## numpy.select() function return an array drawn from elements in choicelist, depending on conditions.
copied_set['total_bedroom_size'] = np.select(cases,room_size)
copied_set
```

```python
copied_set['total_bedroom_size'].value_counts()
```

 Here based on the certain conditions , i have created a new column with named as "total_bedroom_size" and using that cases and conditions categories has been alotted to them by using select function of numpy .
